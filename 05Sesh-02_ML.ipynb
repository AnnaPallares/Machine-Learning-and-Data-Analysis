# Import, adapt and visualize data

````python
from sklearn.datasets import fetch_openml
````

````python
mnist = fetch_openml('mnist_784')
````

````python
X = mnist.data
Y = mnist.target
````

````python
X = X.to_numpy()
Y = Y.to_numpy()
````

````python
# we need to transfor this array of carachters to array of numbers
import numpy as np
Y = np.asarray([eval(i) for i in Y])
````

````python
# negative and positive digits
dn = 0
dp = 1
mask = (Y == dn) | (Y == dp)
X = X[mask, :]
Y = Y[mask]
# the order of assigning first the negative and after the positive is important
# if we don't do i like this, all digits can be assigned to a wrong one
Y[Y == dn] = -1
Y[Y == dp] = +1
````

````python
from random import sample

n = 100
nt = 1000
mask = np.asarray(sample(range(X.shape[0]),n+nt))

X = X[mask,:]
Y = Y[mask]
mask = range(n,n+nt)

Xt = X[mask,]
Yt = Y[mask]
mask = range(0,n)

X = X[mask,]
Y = Y[mask]

print(X.shape)
print(Y.shape)
print(Xt.shape)
print(Yt.shape)
````

````python
from sklearn.linear_model import RidgeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix

grid = {'alpha'  : np.logspace(-4,3,30)}

MS = GridSearchCV(estimator   = RidgeClassifier(),
                  param_grid  = grid,
                  scoring     = 'accuracy',
                  cv          = 10)

H = MS.fit(X,Y)
M = RidgeClassifier(alpha= H.best_params_['alpha'])
M.fit(X,Y)
Yp = M.predict(Xt)

print(np.mean(Yp * Yt <= 0))
print(confusion_matrix(Yt,Yp))
````

````python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix

grid = {'penalty' : ['l1'],
        'C'       : np.logspace(-4,3,30),
        'solver'  : ['liblinear']}

MS = GridSearchCV(estimator   = LogisticRegression(),
                  param_grid  = grid,
                  scoring     = 'accuracy',
                  cv          = 10)

H = MS.fit(X,Y)
M = LogisticRegression(penalty = H.best_params_['penalty'],
                       C       = H.best_params_['C'],
                       solver  = H.best_params_['solver'])
M.fit(X,Y)
Yp = M.predict(Xt)

print(np.mean(Yp * Yt <= 0))
print(confusion_matrix(Yt,Yp))
````

````python
print(M.coef_)
````
