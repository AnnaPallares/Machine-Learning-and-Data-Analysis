### Generate data

````python
# import libraries

import numpy as np
import numpy.matlib as nml
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
````

````python
# random generator (used to obtain the same random value)

np.random.seed(11)
````

````python
# define a funtion

def oracle(x):
  # return x
  # return x**2
  # return np.abs(np.abs(x-.4)-.2)+x/2-.1 # piecewise
  return np.sinc(3*x) # exponential function
````

````python
n = 100
sigma = .01
d = 1
````

````python
x = np.random.rand(n)
y = oracle(x) + sigma*np.random.rand(n)
````

````python
lam = 1
````

````python
from sklearn.linear_model import Ridge
M = Ridge(alpha = lam, fit_intercept = False, solver = 'cholesky')
X = np.asarray(np.asmatrix(x).transpose())
M.fit(X,y)
ns = 1000
xs = np.linspace(0,1,ns)
Xs = np.asarray(np.asmatrix(xs).transpose())
ys = M.predict(Xs)
````

````python
fig, ax= plt.subplots()
ax.plot(X,y,'bo')
ax.plot(Xs,ys,'r')
````

````python
X = np.asmatrix(x).transpose()
yy = np.asmatrix(y).transpose()
w = np.linalg.solve(
    np.matmul(np.transpose(X),X)+lam*np.eye(d),
    np.matmul(np.transpose(X),yy))
````

````python
np.sum(np.abs(w-M.coef_))
````

````python
lam = .1
gam = 10
````

````python
from sklearn.kernel_ridge import KernelRidge
M = KernelRidge(alpha = lam, kernel = 'rbf', gamma = gam)
X = np.asarray(np.asmatrix(x).transpose())
M.fit(X,y)
ns = 1000
xs = np.linspace(0,1,ns)
Xs = np.asarray(np.asmatrix(xs).transpose())
ys = M.predict(Xs)
````

````python
fig, ax = plt.subplots()
ax.plot(X,y,'bo')
ax.plot(Xs,ys,'r')
````
