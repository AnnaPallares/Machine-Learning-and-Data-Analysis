# Import, adapt and visualize data

````python
from sklearn.datasets import fetch_openml
````

````python
mnist = fetch_openml('mnist_784')
````

````python
X = mnist.data
Y = mnist.target
````

````python
X = X.to_numpy()
Y = Y.to_numpy()
````

````python
# we need to normalize the data

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X = scaler.fit_transform(X)
````

````python
# we need to transfor this array of carachters to array of numbers

import numpy as np

Y = np.asarray([eval(i) for i in Y])
````

````python
from random import sample

n = 3000
nt = 10000
mask = np.asarray(sample(range(X.shape[0]),n+nt))

X = X[mask,:]
Y = Y[mask]
mask = range(n,n+nt)

Xt = X[mask,]
Yt = Y[mask]
mask = range(0,n)

X = X[mask,]
Y = Y[mask]
````

````python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix

M = SVC(C      = 1,
        kernel = 'linear')

M.fit(X,Y)
Yp = M.predict(Xt)

print(np.mean(Yp != Yt))
print(confusion_matrix(Yt,Yp))
````

````python
from sklearn.model_selection import GridSearchCV

# if we optimize the hyperparameter, we get even a best result
grid = {'C'      : np.logspace(-4,3,10),
        'kernel' : ['linear']}

MS = GridSearchCV(estimator   = SVC(),
                  param_grid  = grid,
                  scoring     = 'accuracy',
                  cv          = 10,
                  verbose     = 2)

H = MS.fit(X,Y)

M = SVC(C      = H.best_params_['C'],
        kernel = H.best_params_['kernel'])

M.fit(X,Y)
Yp = M.predict(Xt)

print(np.mean(Yp * Yt <= 0))
print(confusion_matrix(Yt,Yp))
````

````python
from sklearn.model_selection import GridSearchCV

# if moreover we optimize the parameter alpha, computational time increases but the model should be even better
grid = {'C'      : np.logspace(-4,3,10),
        'kernel' : ['rbf'],
        'gamma'  : np.logspace(-4,3,10)}

MS = GridSearchCV(estimator   = SVC(),
                  param_grid  = grid,
                  scoring     = 'accuracy',
                  cv          = 10,
                  verbose     = 2)

H = MS.fit(X,Y)

M = SVC(C      = H.best_params_['C'],
        kernel = H.best_params_['kernel'],
        gamma  = H.best_params_['gamma'])

M.fit(X,Y)
Yp = M.predict(Xt)

print(np.mean(Yp * Yt <= 0))
print(confusion_matrix(Yt,Yp))
````
